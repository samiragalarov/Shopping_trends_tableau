````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame
# For example, df = pd.read_csv('your_data.csv')

# List of categorical columns
CatCols = ['Gender', 'Category', 'Size', 'Season']

# Set the style for seaborn
sns.set(style="whitegrid")

# Create subplots for each categorical column
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Flatten the 2D array of subplots for easier indexing
axes = axes.flatten()

# Loop through each categorical column and create count plots
for i, col in enumerate(CatCols):
    sns.countplot(x=col, data=df, ax=axes[i])
    axes[i].set_title(f'{col} Distribution')
    axes[i].tick_params(axis='x')  # Rotate x-axis labels for better visibility

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()

````

## Age distribution

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame
# For example, df = pd.read_csv('your_data.csv')

# List of numeric columns
NumericCols = ['Age', 'Purchase Amount (USD)', 'Review Rating', 'Previous Purchases']

# Set the style for seaborn
sns.set(style="whitegrid")

# Create subplots for each numeric column
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Flatten the 2D array of subplots for easier indexing
axes = axes.flatten()

# Loop through each numeric column and create distribution plots
for i, col in enumerate(NumericCols):
    sns.histplot(df[col], kde=True, ax=axes[i], bins=20)
    axes[i].set_title(f'{col} Distribution')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
````

````python
import matplotlib.pyplot as plt
import pandas as pd

print(df.shape)


# Group by Size and calculate total revenue for each size
size_revenue = df.groupby('Size')['Purchase Amount (USD)'].sum()

# Group by Category and calculate total revenue for each category
category_revenue = df.groupby('Category')['Purchase Amount (USD)'].sum()

# Group by Season and calculate total revenue for each season
season_revenue = df.groupby('Season')['Purchase Amount (USD)'].sum()

# Plotting pie charts
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Pie chart for Size
axes[0].pie(size_revenue, labels=size_revenue.index, autopct='%1.1f%%', startangle=90)
axes[0].set_title('Total Revenue by Size')


# Pie chart for Category
axes[1].pie(category_revenue, labels=category_revenue.index, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Total Revenue by Category')

# Pie chart for Season
axes[2].pie(season_revenue, labels=season_revenue.index, autopct='%1.1f%%', startangle=90)
axes[2].set_title('Total Revenue by Season')

plt.show()
````

````python
df = pd.DataFrame(df)

# Group by 'Subscription Status' and calculate total revenue mean
subscription_status_mean = df.groupby('Subscription Status')['Purchase Amount (USD)'].mean()
print(subscription_status_mean)

# Group by 'Discount Applied' and calculate total revenue mean
discount_applied_mean = df.groupby('Discount Applied')['Purchase Amount (USD)'].mean()
print(discount_applied_mean)
# Group by 'Promo Code Used' and calculate total revenue mean
promo_code_used_mean = df.groupby('Promo Code Used')['Purchase Amount (USD)'].mean()
print(promo_code_used_mean)
````
